```python
import json
```

```python
fd = open("products.json",'r')
r = fd.read()
fd.close()
```

```python
products = json.loads(r)
```

```python
products
```

```python
ui_prod = input("Enter the product Id: ")
ui_quan = int(input("Enter the quantity of the Product: "))
for i in products.keys():
    while ui_prod == i:
        if ui_quan <= products[i]['quan']:
            print("\nWelcome to Biscuits Shop\n")
            print("Product Name:",products[ui_prod]['Name'])
            print("Product Price:",products[ui_prod]['price'])
            print("Expiry date of Product:",products[ui_prod]['expiry'])
            print("Product Brand:",products[ui_prod]['brand'])
            print("------BILLING DETAILS:------- ")
            print("Total Amount: Rs.",products[ui_prod]['price']*ui_quan) 
            print("\n\nThank You. Visit Again")
            products[ui_prod]["quan"] = products[ui_prod]["quan"] - ui_quan
        else:
            print("Sorry. Stock is not avaiable: ",products[ui_prod]['quan'])
        break
if ui_prod not in products.keys():
    print("Sorry. No product found for ID: ",ui_prod,"\Please enter valid product ID")
    
```

```python
js = json.dumps(products)

fd = open("products.json",'w')
fd.write(js)
fd.close()
```

Update items after sold

```python
sales = {1 : {'prod':ui_prod, 'quan':ui_quan, 'amount':products[ui_prod]['price']*ui_quan}}
sal = json.dumps(sales)

fd = open("sales.json",'w')
fd.write(sal)
fd.close()
```

```python
fd = open("sales.json",'r')
trans = fd.read()
fd.close()

sales = json.loads(trans)
num = len(sales)+1

sales[num] = {'prod':ui_prod, 'quann' : ui_quan, 'amount': products[ui_prod]['price'] * ui_quan}
sale = json.dumps(sales)

fd = open("sales.json",'w')
fd.write(sale)
fd.close()
```
